```python
# Abrir 201911_EnvigmuBDD_indicadores_violencia.csv
df_violencia = pd.read_csv('201911_EnvigmuBDD_indicadores_violencia.csv', sep=',', encoding='latin-1')
df_violencia.head()
```

```python
# Quedarme con edad y prov, f1_s2_16_2, f1_s2_7
df_violencia = df_violencia[['id_per','f1_s2_3', 'prov', 'f1_s2_16_1', 'f1_s2_7']]
# Renombrar las columnas
df_violencia.rename(columns={
'f1_s2_3':	'Edad',
'f1_s2_16_1':	'Nivel de instrucción',
'f1_s2_7':	'Estado civil o conyugal ',
}, inplace=True)
```

```python
# JOi de df_depurado y df_violencia con id_per
df_violencia_depurado = pd.merge(df_depurado, df_violencia, on='id_per')
df_violencia_depurado.shape
```

```python
# Nueva variable que sea rango de edad menores de 10 años de 11 a 20 años de 21 a 30 años de 31 a 40 años de 41 a 50 años de 51 a 60 años de 61 a 70 años de 71 a 80 años de 81 a 90 años de 91 a 100 años de más de 100 años
df_violencia_depurado['Rango de edad'] = df_violencia_depurado['Edad'].apply(lambda x:
    'Menores de 10 años' if x < 10 else
    '11 a 20 años' if 11 <= x <= 20 else
    '21 a 30 años' if 21 <= x <= 30 else
    '31 a 40 años' if 31 <= x <= 40 else
    '41 a 50 años' if 41 <= x <= 50 else
    '51 a 60 años' if 51 <= x <= 60 else
    '61 a 70 años' if 61 <= x <= 70 else
    '71 a 80 años' if 71 <= x <= 80 else
    '81 a 90 años' if 81 <= x <= 90 else
    '91 a 100 años' if 91 <= x <= 100 else
    'Más de 100 años' if x > 100 else
    'No se sabe'
)
```
